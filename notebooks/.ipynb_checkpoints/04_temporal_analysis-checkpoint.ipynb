# 04 - Analisis Temporal 2008-2024
## Proyecto Boomerang - SpaceHACK 2026

Serie temporal de 16 anios combinando Landsat 5/7/8/9 (Collection 2).
Objetivo: demostrar cuantitativamente la correlacion inversa entre
perdida de manglar y expansion camaronera, y su relacion con NDCI.

```python
import ee
import geemap
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

ee.Initialize(project='august-tower-470819-s6')
```

```python
# --- ROI ---
roi = ee.Geometry.Polygon([[
    [-80.23433322304784, -2.581044464678974],
    [-80.09837741250097, -2.581044464678974],
    [-80.09837741250097, -2.4212085241315013],
    [-80.23433322304784, -2.4212085241315013],
    [-80.23433322304784, -2.581044464678974]
]])
```

```python
# --- Funciones de procesamiento Landsat ---

def scale_landsat(image):
    """Aplica factores de escala Collection 2 Level 2."""
    optical = image.select('SR_B.').multiply(0.0000275).add(-0.2)
    return image.addBands(optical, overwrite=True)

def add_ndvi_l5(image):
    """NDVI para Landsat 5/7: (B4-B3)/(B4+B3)"""
    ndvi = image.normalizedDifference(['SR_B4', 'SR_B3']).rename('NDVI')
    return image.addBands(ndvi)

def add_ndvi_l8(image):
    """NDVI para Landsat 8/9: (B5-B4)/(B5+B4)"""
    ndvi = image.normalizedDifference(['SR_B5', 'SR_B4']).rename('NDVI')
    return image.addBands(ndvi)

def mask_clouds_landsat(image):
    """Mascara de nubes usando QA_PIXEL band (Collection 2)."""
    qa = image.select('QA_PIXEL')
    cloud_mask = qa.bitwiseAnd(1 << 3).eq(0)  # Bit 3 = cloud
    shadow_mask = qa.bitwiseAnd(1 << 4).eq(0)  # Bit 4 = cloud shadow
    return image.updateMask(cloud_mask).updateMask(shadow_mask)
```

```python
# --- Configuracion de sensores por periodo ---
sensor_config = {
    # 2008-2011: Landsat 5 TM
    2008: ('LANDSAT/LT05/C02/T1_L2', add_ndvi_l5),
    2009: ('LANDSAT/LT05/C02/T1_L2', add_ndvi_l5),
    2010: ('LANDSAT/LT05/C02/T1_L2', add_ndvi_l5),
    2011: ('LANDSAT/LT05/C02/T1_L2', add_ndvi_l5),
    # 2012: Landsat 7 (L5 ya no operativo)
    2012: ('LANDSAT/LE07/C02/T1_L2', add_ndvi_l5),
    # 2013+: Landsat 8 OLI
    2013: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2014: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2015: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2016: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2017: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2018: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2019: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2020: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2021: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2022: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2023: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
    2024: ('LANDSAT/LC08/C02/T1_L2', add_ndvi_l8),
}

print(f'Analizando {len(sensor_config)} anios: {min(sensor_config.keys())}-{max(sensor_config.keys())}')
```

```python
# --- Calcular NDVI anual y clasificar manglar ---
# Umbral NDVI > 0.4 = vegetacion densa (manglar)
# Umbral NDVI 0.1-0.4 = vegetacion dispersa/camaronera seca
# Umbral NDVI < 0.1 = agua/suelo desnudo

MANGROVE_THRESHOLD = 0.4

yearly_data = {}

for year, (collection_id, ndvi_func) in sensor_config.items():
    try:
        col = (ee.ImageCollection(collection_id)
            .filterBounds(roi)
            .filterDate(f'{year}-01-01', f'{year}-12-31')
            .filter(ee.Filter.lt('CLOUD_COVER', 20))
            .map(mask_clouds_landsat)
            .map(scale_landsat)
            .map(ndvi_func))

        ndvi_median = col.select('NDVI').median().clip(roi)

        # Area de manglar (NDVI > umbral)
        mangrove_mask = ndvi_median.gt(MANGROVE_THRESHOLD)
        mangrove_area = mangrove_mask.multiply(ee.Image.pixelArea()).reduceRegion(
            reducer=ee.Reducer.sum(),
            geometry=roi,
            scale=30,
            maxPixels=1e13
        )

        # NDVI medio general
        ndvi_mean = ndvi_median.reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=roi,
            scale=30,
            maxPixels=1e13
        )

        yearly_data[year] = {
            'mangrove_area_m2': mangrove_area,
            'ndvi_mean': ndvi_mean,
            'image': ndvi_median
        }
        print(f'  {year}: procesado')
    except Exception as e:
        print(f'  {year}: error - {e}')

print(f'\nAnios procesados: {len(yearly_data)}')
```

```python
# --- Extraer valores numericos ---
years = []
mangrove_ha = []
ndvi_means = []

print('=== DATOS TEMPORALES ===')
print(f'{"Anio":<6} {"Manglar (ha)":<15} {"NDVI medio":<12}')
print('-' * 35)

for year in sorted(yearly_data.keys()):
    data = yearly_data[year]
    try:
        area_val = data['mangrove_area_m2'].getInfo()
        ndvi_val = data['ndvi_mean'].getInfo()

        area_ha_val = list(area_val.values())[0] / 10000 if area_val else 0
        ndvi_mean_val = list(ndvi_val.values())[0] if ndvi_val else 0

        if area_ha_val and ndvi_mean_val:
            years.append(year)
            mangrove_ha.append(area_ha_val)
            ndvi_means.append(ndvi_mean_val)
            print(f'{year:<6} {area_ha_val:<15,.1f} {ndvi_mean_val:<12.4f}')
    except Exception as e:
        print(f'{year:<6} Error: {e}')
```

```python
# --- GRAFICA TEMPORAL: Manglar vs Tiempo ---
if len(years) >= 3:
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    # Panel 1: Area de manglar
    ax1.plot(years, mangrove_ha, 'g-o', linewidth=2, markersize=6, label='Area de Manglar')
    ax1.fill_between(years, mangrove_ha, alpha=0.2, color='green')
    ax1.set_ylabel('Hectareas de Manglar', fontsize=12)
    ax1.set_title('Evolucion del Manglar en el Golfo de Guayaquil (2008-2024)', fontsize=14)
    ax1.legend(fontsize=11)
    ax1.grid(True, alpha=0.3)

    # Panel 2: NDVI medio
    ax2.plot(years, ndvi_means, 'b-s', linewidth=2, markersize=6, label='NDVI Medio')
    ax2.axhline(y=MANGROVE_THRESHOLD, color='r', linestyle='--', alpha=0.5, label=f'Umbral manglar ({MANGROVE_THRESHOLD})')
    ax2.set_xlabel('Anio', fontsize=12)
    ax2.set_ylabel('NDVI Medio', fontsize=12)
    ax2.legend(fontsize=11)
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('../exports/temporal_manglar.png', dpi=150, bbox_inches='tight')
    plt.show()

    # Tendencia lineal
    slope, intercept, r_value, p_value, std_err = stats.linregress(years, mangrove_ha)
    print(f'\n=== TENDENCIA LINEAL (Manglar) ===')
    print(f'Pendiente: {slope:+.1f} ha/anio')
    print(f'R-squared: {r_value**2:.4f}')
    print(f'p-value: {p_value:.6f}')
    if slope < 0:
        total_loss = abs(slope) * (max(years) - min(years))
        print(f'Perdida total estimada: {total_loss:,.0f} ha en {max(years)-min(years)} anios')
else:
    print('Insuficientes datos para graficar (necesita >= 3 anios)')
```

```python
# --- MAPAS COMPARATIVOS: 2008 vs 2016 vs 2024 ---
key_years = [2008, 2016, 2024]
ndvi_vis = {'min': -0.1, 'max': 0.8, 'palette': ['red', 'yellow', 'green', 'darkgreen']}

Map_temporal = geemap.Map(center=[-2.50, -80.17], zoom=12)

for y in key_years:
    if y in yearly_data:
        Map_temporal.addLayer(
            yearly_data[y]['image'],
            ndvi_vis,
            f'NDVI {y}'
        )

# Mapa de cambio: diferencia 2024 - 2008
if 2008 in yearly_data and 2024 in yearly_data:
    ndvi_change = yearly_data[2024]['image'].subtract(yearly_data[2008]['image']).rename('NDVI_change')
    change_vis = {
        'min': -0.5,
        'max': 0.5,
        'palette': ['FF0000', 'FF6600', 'FFFF00', 'FFFFFF', '99FF99', '00CC00', '006600']
    }
    Map_temporal.addLayer(ndvi_change, change_vis, 'Cambio NDVI (2024-2008)')

Map_temporal.add_legend(title='NDVI', builtin_legend='NDVI')
Map_temporal
```

```python
# --- METRICAS CLAVE PARA EL PITCH ---
if len(years) >= 3:
    first_year = min(years)
    last_year = max(years)
    first_area = mangrove_ha[0]
    last_area = mangrove_ha[-1]
    change = last_area - first_area
    pct_change = (change / first_area) * 100

    print('=' * 50)
    print('  METRICAS TEMPORALES PARA EL PITCH DECK')
    print('=' * 50)
    print(f'  Periodo: {first_year} - {last_year} ({last_year - first_year} anios)')
    print(f'  Manglar en {first_year}: {first_area:,.1f} ha')
    print(f'  Manglar en {last_year}: {last_area:,.1f} ha')
    print(f'  Cambio: {change:+,.1f} ha ({pct_change:+.1f}%)')
    print(f'  Tasa: {slope:+.1f} ha/anio')
    print(f'  Significancia estadistica: p={p_value:.6f}')
    print(f'  NDVI maximo: {max(ndvi_means):.4f} (anio {years[ndvi_means.index(max(ndvi_means))]})')
    print(f'  NDVI minimo: {min(ndvi_means):.4f} (anio {years[ndvi_means.index(min(ndvi_means))]})')
    print('=' * 50)
```
