Genera todas las visualizaciones y metricas necesarias para las slides.

```python
import ee
import geemap
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import os

ee.Initialize(project='august-tower-470819-s6')

os.makedirs('../exports', exist_ok=True)
print('Directorio exports/ creado')
```

```python
# --- ROI ---
roi = ee.Geometry.Polygon([[
    [-80.23433322304784, -2.581044464678974],
    [-80.09837741250097, -2.581044464678974],
    [-80.09837741250097, -2.4212085241315013],
    [-80.23433322304784, -2.4212085241315013],
    [-80.23433322304784, -2.581044464678974]
]])

# Sentinel-2 2024
s2 = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterBounds(roi)
    .filterDate('2024-01-01', '2024-12-31')
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))
    .median()
    .clip(roi))
```

```python
# --- EXPORTAR MAPA RGB ---
Map_rgb = geemap.Map(center=[-2.50, -80.17], zoom=13)
Map_rgb.addLayer(s2, {'bands': ['B4', 'B3', 'B2'], 'max': 3000}, 'RGB')

# Exportar como imagen estatica
geemap.get_image_thumbnail(
    s2,
    roi,
    vis_params={'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 3000},
    dimensions=1024,
    out_img='../exports/rgb_sentinel2_2024.png'
)
print('Exportado: rgb_sentinel2_2024.png')
```

```python
# --- EXPORTAR NDVI ---
ndvi = s2.normalizedDifference(['B8', 'B4']).rename('NDVI')

geemap.get_image_thumbnail(
    ndvi,
    roi,
    vis_params={'min': -0.1, 'max': 0.8, 'palette': ['red', 'yellow', 'green', 'darkgreen']},
    dimensions=1024,
    out_img='../exports/ndvi_2024.png'
)
print('Exportado: ndvi_2024.png')
```

```python
# --- EXPORTAR HEATMAP NDCI ---
mndwi = s2.normalizedDifference(['B3', 'B11']).rename('MNDWI')
water_mask = mndwi.gt(0)
ndci = s2.normalizedDifference(['B5', 'B4']).rename('NDCI')
ndci_water = ndci.updateMask(water_mask)

geemap.get_image_thumbnail(
    ndci_water,
    roi,
    vis_params={'min': -0.1, 'max': 0.2, 'palette': ['0000FF', '00FFFF', 'FFFF00', 'FF0000']},
    dimensions=1024,
    out_img='../exports/ndci_heatmap_2024.png'
)
print('Exportado: ndci_heatmap_2024.png')
```

```python
# --- PANEL DE METRICAS DE IMPACTO ---
# Estos valores se llenan con los resultados reales de notebooks 02, 04 y 05
# Por ahora son placeholders que se actualizan al correr la pipeline completa

metrics = {
    'Manglar perdido (2008-2024)': 'X,XXX ha',
    'Camaronera nueva (2008-2024)': 'X,XXX ha',
    'Correlacion Pearson (manglar vs NDCI)': 'r = -0.XX',
    'Overall Accuracy (clasificador)': 'XX.X%',
    'Kappa de Cohen': '0.XX',
    'Area en riesgo critico BRI': 'X,XXX ha',
    'Valor ecosistemico en riesgo': '$XXX M/anio',
    'Frecuencia de monitoreo': '5 dias (Sentinel-2)',
}

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')
ax.set_title('METRICAS DE IMPACTO - Proyecto Boomerang', fontsize=16, fontweight='bold', pad=20)

y_pos = 0.9
for metric, value in metrics.items():
    ax.text(0.05, y_pos, metric, fontsize=12, transform=ax.transAxes, va='top')
    ax.text(0.95, y_pos, value, fontsize=12, fontweight='bold', transform=ax.transAxes, va='top', ha='right')
    y_pos -= 0.11

plt.tight_layout()
plt.savefig('../exports/impact_metrics.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print('Exportado: impact_metrics.png')
```

```python
# --- DIAGRAMA DEL EFECTO BOOMERANG ---
fig, ax = plt.subplots(figsize=(12, 4))
ax.axis('off')

steps = [
    ('Expansion\nCamaronera', '#FF4500'),
    ('Deforestacion\nManglar', '#8B4513'),
    ('Perdida Filtro\nBiologico', '#DAA520'),
    ('Aumento\nMarea Roja', '#DC143C'),
    ('Hipoxia en\nPiscinas', '#800000'),
    ('Perdida de\nProduccion', '#4B0082'),
]

for i, (label, color) in enumerate(steps):
    x = 0.08 + i * 0.155
    circle = plt.Circle((x, 0.5), 0.08, transform=ax.transAxes, color=color, alpha=0.8)
    ax.add_patch(circle)
    ax.text(x, 0.5, label, ha='center', va='center', fontsize=8,
            color='white', fontweight='bold', transform=ax.transAxes)
    if i < len(steps) - 1:
        ax.annotate('', xy=(x + 0.1, 0.5), xytext=(x + 0.055, 0.5),
                    xycoords='axes fraction', textcoords='axes fraction',
                    arrowprops=dict(arrowstyle='->', color='gray', lw=2))

# Flecha de retorno (boomerang)
ax.annotate('', xy=(0.08, 0.25), xytext=(0.855, 0.25),
            xycoords='axes fraction', textcoords='axes fraction',
            arrowprops=dict(arrowstyle='->', color='red', lw=3, linestyle='dashed'))
ax.text(0.47, 0.15, 'EFECTO BOOMERANG', ha='center', fontsize=12,
        color='red', fontweight='bold', transform=ax.transAxes)

ax.set_title('El Ciclo del Efecto Boomerang', fontsize=14, fontweight='bold')
plt.savefig('../exports/boomerang_cycle.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
print('Exportado: boomerang_cycle.png')
```
